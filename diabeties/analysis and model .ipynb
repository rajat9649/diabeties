```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calculate_vif_(X):

    '''X - pandas dataframe'''
    thresh = 5.0
    variables = range(X.shape[1])

    for i in np.arange(0, len(variables)):
        vif = [variance_inflation_factor(X[variables].values, ix) for ix in range(X[variables].shape[1])]
        print(vif)
        maxloc = vif.index(max(vif))
        if max(vif) > thresh:
            print('dropping \'' + X[variables].columns[maxloc] + '\' at index: ' + str(maxloc))
            del variables[maxloc]

    print('Remaining variables:')
    print(X.columns[variables])
    return X
```

```python
df=df[['Glucose','Insulin','Pregnancies','BMI','Outcome']]
```

```python
from sklearn.model_selection import train_test_split
train,test=train_test_split(df,test_size=0.2,random_state=42,shuffle=True)
```

```python
from sklearn.linear_model import LogisticRegression 
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
```

```python
from sklearn.feature_selection import RFECV
#lr=LogisticRegression(random_state=0)
rfe=RandomForestClassifier(n_estimators=600,random_state=0)
dtc=DecisionTreeClassifier(random_state=0)
#xb=XGBClassifier(random_state=0)
```

```python
#result_1=RFECV(lr,step=1,cv=6).fit(train.iloc[:,:-1],train.iloc[:,-1])
result_2=RFECV(rfe,step=1,cv=6).fit(train.iloc[:,:-1],train.iloc[:,-1])
result_3=RFECV(dtc,step=1,cv=6).fit(train.iloc[:,:-1],train.iloc[:,-1])
#result_4=RFECV(xb,step=1,cv=6).fit(train.iloc[:,:-1],train.iloc[:,-1])
```

```python
#print('logisticregression',np.array(df.columns)[:-1],result_1.support_)
print('randomforest',np.array(df.columns)[:-1],result_2.support_)
print('decsiontree',np.array(df.columns)[:-1],result_3.support_)
#print('xgboost',np.array(df.columns)[:-1],result_4.support_)
```

```python
print('random forest score according to rfe')
rfe=rfe.fit(train.iloc[:,:-1],train.iloc[:,-1])
rfe.score(train.iloc[:,:-1],train.iloc[:,-1])
```

```python
print('descision tree score according to rfe')
dtc.fit(train.iloc[:,:-1],train.iloc[:,-1])
dtc.score(train.iloc[:,:-1],train.iloc[:,-1])
```

```python
from sklearn.metrics import accuracy_score
y_predict=rfe.predict(test.iloc[:,:-1])
accuracy_score(y_predict,test.iloc[:,-1])
```

```python
from sklearn.model_selection import validation_curve
param_range = np.arange(1, 250, 2)
train_scores, test_scores = validation_curve(RandomForestClassifier(), 
                                             train.iloc[:,:-1], 
                                             train.iloc[:,-1], 
                                             param_name="n_estimators", 
                                             param_range=param_range,
                                             cv=6, 
                                             scoring="accuracy", 
                                             n_jobs=-1)


train_mean = np.mean(train_scores, axis=1)
train_std = np.std(train_scores, axis=1)

test_mean = np.mean(test_scores, axis=1)
test_std = np.std(test_scores, axis=1)

plt.plot(param_range, train_mean, label="Training score", color="black")
plt.plot(param_range, test_mean, label="Cross-validation score", color="dimgrey")

plt.title("Validation Curve With Random Forest")
plt.xlabel("Number Of Trees")
plt.ylabel("Accuracy Score")
plt.tight_layout()
plt.legend(loc="best")
plt.show()
```
